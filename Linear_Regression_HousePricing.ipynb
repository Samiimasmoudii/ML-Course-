```python
house_data_raw = pd.read_csv('house.csv')
house_data = house_data_raw[house_data_raw['loyer']<7000]
```

```python
plt.plot(house_data['surface'],house_data['loyer'],'ro')
plt.show()
```

```python
plt.plot(house_data['surface'],house_data['loyer'],'ro',markersize=2)
plt.show()
```

```python
surface=house_data['surface'].values
loyer=house_data['loyer'].values
x = np.column_stack((np.ones(len(surface)), surface))

# Convert `y` (loyer) to a NumPy array
y = np.array(loyer)

# Calculate theta using the normal equation
theta = np.linalg.inv(x.T.dot(x)).dot(x.T).dot(y)


```

```python
import random
print('Theta 0 =',theta[0])
print('Theta 1 =',theta[1])
print("L'equation est Y^i = ",theta[0],"+",theta[1]," * Xi")
xtest=random.randint(50,200)
print("pour X = ",xtest, "y =",theta[0]+theta[1]*xtest )
xtest=random.randint(50,200)
print("pour X = ",xtest, "y =",theta[0]+theta[1]*xtest )
xtest=random.randint(50,200)
print("pour X = ",xtest, "y =",theta[0]+theta[1]*xtest )
```

```python
import matplotlib.pyplot as plt

# Tracer le nuage de points
plt.scatter(surface, loyer, color='blue', label='DonnÃ©es rÃ©elles')

# Tracer la droite de rÃ©gression
x_vals = np.array([min(surface), max(surface)])  # Points pour la ligne droite (x1 et x2)
y_vals = theta[0] + theta[1] * x_vals  # Calcul des valeurs correspondantes de y (y1 et y2)
plt.plot(x_vals, y_vals, color='red', label='RÃ©gression linÃ©aire',markersize=0.5)

# Ajouter des labels et une lÃ©gende
plt.xlabel('Surface (mÂ²)')
plt.ylabel('Loyer (TND)')
plt.title('RÃ©gression linÃ©aire - Loyer vs Surface')
plt.legend()

# Afficher la figure
plt.show()
```
